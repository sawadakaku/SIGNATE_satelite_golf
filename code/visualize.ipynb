```python
def f(im):
    im = ((im * 255) / np.max(im)).astype(np.uint8)
    l = []
    for i in range(im.shape[2]):
        l.append(cv2.adaptiveThreshold(im[:, :, i], 1, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 2)[:, :, np.newaxis])
    return np.concatenate(l, axis=2)
```
